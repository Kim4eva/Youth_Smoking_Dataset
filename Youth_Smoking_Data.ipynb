```python
from scipy import stats

filtered_df = df[(df['Year'] >= 2020) & (df['Year'] <= 2024)]

avg_smoking_with_counseling = filtered_df[filtered_df['Access_to_Counseling'] == 'Yes'].groupby('Year')['Smoking_Prevalence'].mean()
avg_smoking_without_counseling = filtered_df[filtered_df['Access_to_Counseling'] == 'No'].groupby('Year')['Smoking_Prevalence'].mean()

smoking_trends = pd.DataFrame({
    'With Counseling': avg_smoking_with_counseling,
    'Without Counseling': avg_smoking_without_counseling
})


plt.figure(figsize=(12, 6))
smoking_trends.plot(kind='line', marker='o')
plt.title('Smoking Rates Over Time by Access to Counseling')
plt.xlabel('Year')
plt.ylabel('Average Smoking Rate (%)')
plt.xticks(smoking_trends.index)
plt.legend(title='Access to Counseling')
plt.show()


t_stat, p_value = stats.ttest_ind(
    filtered_df[filtered_df['Access_to_Counseling'] == 'Yes']['Smoking_Prevalence'],
    filtered_df[filtered_df['Access_to_Counseling'] == 'No']['Smoking_Prevalence'],
    equal_var=False
)

print(f'T-test Statistic: {t_stat:.2f}, P-value: {p_value:.4f}')
```

```python
filtered_df = df[(df['Year'] >= 2020) & (df['Year'] <= 2024)]

avg_smoking_by_supervision = filtered_df.groupby(['Year', 'Parental_Supervision'])['Smoking_Prevalence'].mean().unstack()

avg_supervision_over_time = filtered_df.groupby('Year')['Parental_Supervision'].mean()

plt.figure(figsize=(12, 6))
avg_smoking_by_supervision.plot(kind='line', marker='o')
plt.title('Smoking Rates Over Time by Parental Supervision Level')
plt.xlabel('Year')
plt.ylabel('Average Smoking Prevalence (%)')
plt.xticks(avg_smoking_by_supervision.index)  # Ensure all years are shown on the x-axis
plt.legend(title='Supervision')
plt.show()

# ANOVA to check if there are significant differences in smoking rates by parental supervision level
anova_results = stats.f_oneway(
    *[group['Smoking_Prevalence'].values for name, group in filtered_df.groupby('Parental_Supervision')]
)

print(f'ANOVA F-statistic: {anova_results.statistic:.2f}, P-value: {anova_results.pvalue:.4f}')
```

```python
import statsmodels.api as sm

# Regression analysis for Smoking Prevalence
X = df['Year']
y = df['Smoking_Prevalence']

# Adding a constant to the predictor (intercept)
X = sm.add_constant(X)

# regression model
model = sm.OLS(y, X).fit()

print(model.summary())


y_drug = df['Drug_Experimentation']
model_drug = sm.OLS(y_drug, X).fit()
print(model_drug.summary())
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


X = df[['Year']].copy()  
X.loc[:, 'Intercept'] = 1 

vif_data = pd.DataFrame()
vif_data['Variable'] = X.columns
vif_data['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]

print(vif_data)
```

```python
from scipy.stats import spearmanr

corr, p_value = spearmanr(df['Year'], df['Smoking_Prevalence'])
print(f"Spearman's correlation: {corr}, P-value: {p_value}")
```

```python
from scipy.stats import kruskal

groups = [group['Smoking_Prevalence'].values for name, group in df.groupby('Year')]

stat, p_value = kruskal(*groups)
print(f'Kruskal-Wallis test statistic: {stat}, P-value: {p_value}')
```
